# Customer Churn Prediction - Model Testing

This notebook tests the saved best model pipeline with sample input data.


```python
#Import Libraries
import pandas as pd
import pickle
```

```python
#Load Saved Model
with open("../models/best_model_pipeline.pkl", "rb") as f:
    model_pipeline = pickle.load(f)

with open("../models/feature_columns.pkl", "rb") as f:
    feature_columns = pickle.load(f)
```

```python
#Sample Input
sample_data = pd.DataFrame([{
    'Age': 30,
    'Gender': 'Male',
    'Income': 50000,
    'SpendingScore': 60,
    'PurchaseAmount': 2000,
    'ProductCategory': 'Electronics',
    'PaymentMethod': 'UPI',
    'City': 'Bangalore',
    'State': 'Karnataka',
    'Country': 'India',
    'IsActive': 'Yes',
    'Returns': 1,
    'DiscountUsed': 10,
    'ReviewScore': 4,
    'Browser': 'Chrome',
    'Device': 'Mobile',
    'SessionTime': 120,
    'DaysSinceLastPurchase': 20
}])

sample_data = sample_data[feature_columns]
sample_data
```

```python
#Prediction
prediction = model_pipeline.predict(sample_data)[0]
probability = model_pipeline.predict_proba(sample_data)[0][1]

print("Prediction:", prediction)
print(f"Churn Probability: {probability*100:.2f}%")
```

```python
# Final Result
if prediction == 1:
    print("Customer is likely to CHURN")
else:
    print("Customer is likely to STAY")
```
