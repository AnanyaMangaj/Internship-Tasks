# Customer Churn Prediction - Exploratory Data Analysis

This notebook explores:
- Churn distribution
- Customer behavior patterns
- Income and spending relationships
- Product and engagement trends
- Correlation patterns


```python
#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style("whitegrid")
```

```python
#Load Cleaned Data
df = pd.read_csv("../data/processed/cleaned_customer_churn.csv")
df.head()
```

```python
#Churn Distribution
plt.figure(figsize=(6,4))
sns.countplot(x='Churn', data=df)
plt.title("Customer Churn Distribution")
plt.show()
```

```python
#Churn Percentage
df['Churn'].value_counts(normalize=True) * 100
```

```python
#Gender vs Churn
plt.figure(figsize=(6,4))
sns.countplot(x='Gender', hue='Churn', data=df)
plt.title("Gender vs Churn")
plt.show()
```

```python
#IsActive vs Churn
plt.figure(figsize=(6,4))
sns.countplot(x='IsActive', hue='Churn', data=df)
plt.title("Active Status vs Churn")
plt.show()
```

```python
#ProductCategory vs Churn
plt.figure(figsize=(10,5))
sns.countplot(x='ProductCategory', hue='Churn', data=df)
plt.xticks(rotation=45)
plt.title("Product Category vs Churn")
plt.show()
```

```python
#Income vs Churn
plt.figure(figsize=(8,5))
sns.boxplot(x='Churn', y='Income', data=df)
plt.title("Income vs Churn")
plt.show()
```

```python
#SpendingScore vs Churn
plt.figure(figsize=(8,5))
sns.boxplot(x='Churn', y='SpendingScore', data=df)
plt.title("Spending Score vs Churn")
plt.show()
```

```python
#PurchaseAmount vs Churn
plt.figure(figsize=(8,5))
sns.boxplot(x='Churn', y='PurchaseAmount', data=df)
plt.title("Purchase Amount vs Churn")
plt.show()
```

```python
#DaysSinceLastPurchase vs Churn
plt.figure(figsize=(8,5))
sns.boxplot(x='Churn', y='DaysSinceLastPurchase', data=df)
plt.title("Days Since Last Purchase vs Churn")
plt.show()
```

```python
#Correlation Heatmap
plt.figure(figsize=(12,8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

## EDA Insights

### Key Findings:
- Inactive customers show a higher tendency to churn.
- Spending behavior appears related to churn probability.
- Days since last purchase is likely a strong churn indicator.
- Product category and engagement variables contribute to churn behavior.
- Customer behavioral data is useful for churn prediction.
